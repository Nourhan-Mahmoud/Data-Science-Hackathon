# Load Text File

```python
with open('Skill_BOK.txt', 'r', encoding= 'utf-16', errors = 'ignore') as f:
    c = f.read()
    f.close()
```

# Data Cleaning

## Make All Letters LowerCase

```python
c = c.lower()
```

```python
c
```

## Remove '\n' From All Text

```python
import re

txt = re.sub(r'\n','',c)
```

```python
txt
```

## Remove E-Mails

```python
tt = re.sub(r'[A-Za-z0-9]*@[A-Za-z]*\.?[A-Za-z0-9]*',"", txt)
```

```python
tt
```

## Remove Punctuations

```python
#import string
    
#tt = tt.translate(str.maktrans('','',string.punctuation))

tt = re.sub(r'[^\w\s]',' ', tt)
```

```python
tt
```

## Remove All Numbers

```python
txt = re.sub(r"[0-9]","",tt)
```

```python
txt
```

## Remove Non-English Letters

```python
def remove_arabic(ttext):
    import re
    ttext = re.sub("[^a-zA-Z\,\.\+]+"," ",ttext)
    return ttext
```

```python
ttt = remove_arabic(txt)
```
