```python
import pandas as pd

df = pd.read_csv('/content/full_qcsv')

df.head()
```

```python
df.drop(columns =["Unnamed: 0"], inplace = True)
```

```python
df.head()
```

```python
paragraphs = df["Paragraph"]
questions = df["Questions"]
```

```python
from allennlp.predictors.predictor import Predictor
import allennlp_models.rc

predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/transformer-qa.2021-02-11.tar.gz")
```

```python
answers = []
for i in range(df.shape[0]):
    x = predictor.predict( passage = paragraphs[i] , question = questions[i])
    answers.append(x["best_span_str"])
```

```python
df["Answer"] = answers
```

```python
df.head()
```

```python
df.to_csv("[Q-A]MachineLearning")
```
